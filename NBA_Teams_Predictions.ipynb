```python
# Define the conference mapping for the teams
eastern_conference = [
    'Atlanta Hawks', 'Boston Celtics', 'Brooklyn Nets', 'Charlotte Hornets',
    'Chicago Bulls', 'Cleveland Cavaliers', 'Detroit Pistons', 'Indiana Pacers',
    'Miami Heat', 'Milwaukee Bucks', 'New York Knicks', 'Orlando Magic',
    'Philadelphia 76ers', 'Toronto Raptors', 'Washington Wizards'
]

western_conference = [
    'Dallas Mavericks', 'Denver Nuggets', 'Golden State Warriors', 'Houston Rockets',
    'LA Clippers', 'Los Angeles Lakers', 'Memphis Grizzlies', 'Minnesota Timberwolves',
    'New Orleans Pelicans', 'Oklahoma City Thunder', 'Phoenix Suns', 'Portland Trail Blazers',
    'Sacramento Kings', 'San Antonio Spurs', 'Utah Jazz'
]
```

```python
# Function to classify teams into conferences
def classify_conference(team_name):
    if team_name in eastern_conference:
        return 'Eastern'
    elif team_name in western_conference:
        return 'Western'
    else:
        return 'Unknown'  # In case any team is not in either list
```

#### Assign playoff qualification based on W/L%

```python
# Define a function to assign playoff qualification based on W/L%
def assign_playoff_qualification(win_percentage):
    if win_percentage >= 0.50:
        return 1  # Team qualifies for playoffs
    else:
        return 0  # Team does not qualify for playoffs

# Apply the function to each row in the 'W/L%' column
merged_df['playoff_qualified'] = merged_df['W_PCT'].apply(assign_playoff_qualification)
```

```python
# Set a threshold and remove highly correlated features
def correlation_filter(data, threshold=0.9):
    corr_matrix = data.corr().abs()  # Get absolute correlations
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))  # Upper triangle
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    return data.drop(columns=to_drop)
```

```python
# Drop highly correlated features with correlation > 0.9 (can adjust threshold)
merged_df_filtered = correlation_filter(merged_df.drop(columns=['playoff_qualified']))
```

```python
# Split data into features (X) and target (y)
X = merged_df_filtered  # Filtered data after correlation-based selection
y = merged_df['playoff_qualified']  # Target variable
```

### Final Winner

```python
# Now that we have the conference winners, we can simulate the finals (Best-of-7 Series)
# Assume the winner is determined by who wins 4 out of 7 games based on model predictions

def simulate_finals(east_winner, west_winner):
    east_wins = 0
    west_wins = 0
    for i in range(7):
        if np.random.rand() > 0.6:  # Random factor for winner in each game
            east_wins += 1
        else:
            west_wins += 1
        if east_wins == 4:
            return "East Conference Winner"
        elif west_wins == 4:
            return "West Conference Winner"
    return "Game 7 Result"

finals_result = simulate_finals(east_winner, west_winner)
print("Finals Result:", finals_result)
```
