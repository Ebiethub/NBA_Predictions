#### Creating Fictional Team of 15 players

```python
# Create a fictional team by randomly selecting 15 players
fictional_team = df.sample(15, random_state=42)
```

```python
fictional_team.head()
```

#### Assign playoff qualification based on W/L%

```python
# Define a function to assign playoff qualification based on W/L%
def assign_playoff_qualification(win_percentage):
    if win_percentage >= 0.50:
        return 1  # Team qualifies for playoffs
    else:
        return 0  # Team does not qualify for playoffs

# Apply the function to each row in the 'W/L%' column
fictional_team['playoff_qualified'] = fictional_team['W_PCT'].apply(assign_playoff_qualification)
```

```python
fictional_team.head()
```

```python
# Dropping some columns

columns_to_drop = ['PLAYER_ID', 'PLAYER_NAME', 'NICKNAME', 'TEAM_ID', 'TEAM_ABBREVIATION',
                   'GP_RANK', 'W_RANK', 'L_RANK', 'W_PCT_RANK', 'MIN_RANK', 
                   'E_OFF_RATING_RANK', 'OFF_RATING_RANK', 'sp_work_OFF_RATING_RANK', 
                   'E_DEF_RATING_RANK', 'DEF_RATING_RANK', 'sp_work_DEF_RATING_RANK', 
                   'E_NET_RATING_RANK', 'NET_RATING_RANK', 'sp_work_NET_RATING_RANK', 
                   'AST_PCT_RANK', 'AST_TO_RANK', 'AST_RATIO_RANK', 'OREB_PCT_RANK', 
                   'DREB_PCT_RANK', 'REB_PCT_RANK', 'TM_TOV_PCT_RANK', 'E_TOV_PCT_RANK', 
                   'EFG_PCT_RANK', 'TS_PCT_RANK', 'USG_PCT_RANK', 'E_USG_PCT_RANK', 
                   'E_PACE_RANK', 'PACE_RANK', 'sp_work_PACE_RANK', 'PIE_RANK', 
                   'FGM_RANK', 'FGA_RANK', 'FGM_PG_RANK', 'FGA_PG_RANK', 'FG_PCT_RANK', 
                   'season_id']

# Filter out the columns that are actually present in the DataFrame
columns_to_drop = [col for col in columns_to_drop if col in fictional_team.columns]

# Drop the columns that exist in the DataFrame
fictional_team = fictional_team.drop(columns=columns_to_drop, axis=1)

# Check the resulting DataFrame
fictional_team = fictional_team.reset_index(drop=True) # Resetting index

fictional_team
```

```python
fictional_team.info()
```

```python
# Calculate correlation matrix
fictional_team_cor = fictional_team.corr()
fictional_team_cor
```

```python
# Plotting the heatmap to visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(fictional_team_cor, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Set a threshold and remove highly correlated features
def correlation_filter(data, threshold=0.9):
    corr_matrix = data.corr().abs()  # Get absolute correlations
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))  # Upper triangle
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    return data.drop(columns=to_drop)
```

```python
# Drop highly correlated features with correlation > 0.9 (can adjust threshold)
fictional_team_filtered = correlation_filter(fictional_team.drop(columns=['playoff_qualified']))
```

```python
fictional_team_filtered
```

```python
# Split data into features (X) and target (y)
X = fictional_team_filtered  # Filtered data after correlation-based selection
y = fictional_team['playoff_qualified']  # Target variable
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
import matplotlib.pyplot as plt
```

```python
# Feature selection using Random Forest importance
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Fit the model
clf.fit(X_train, y_train)
```

```python
# Get feature importances
importances = clf.feature_importances_
feature_names = X.columns
```

```python
# Create a dataframe to display feature importances
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
```

```python
# Plot the feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title("Feature Importance - Random Forest")
plt.show()
```

```python
# Select top K important features (you can set K based on your preference)
K = 10  # Top 10 features
top_features = feature_importance_df['Feature'].head(K).values
X_selected = X[top_features]
```

```python
X_selected.head()
```

```python
# Scale the data for better performance
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_selected_scaled = scaler.fit_transform(X_selected)
```

```python
pd.DataFrame(X_selected_scaled)
```

```python
# Split into training and testing data
X1_train, X1_test, y1_train, y1_test = train_test_split(X_selected_scaled, y, test_size=0.2, random_state=42)
```

```python
# Model Training
# Using RandomForestClassifier to predict whether a team qualifies for the playoffs
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X1_train, y1_train)
```

```python
# Predict playoff qualification
y_pred = clf.predict(X1_test)
```

```python
# Evaluate model performance
from sklearn.metrics import classification_report

print("Playoff Qualification Prediction Report:")
print(classification_report(y1_test, y_pred))
```
