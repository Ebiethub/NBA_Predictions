## Traditional and Advance Teams Stats

### Getting the traditional stats information for team

```python
# traditional stats

raw_api_url = 'https://stats.nba.com/stats/leaguedashteamstats?Conference=&DateFrom=&DateTo=&Division=&GameScope=&GameSegment=&Height=&ISTRound=&LastNGames=0&LeagueID=00&Location=&MeasureType=Base&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season=2023-24&SeasonSegment=&SeasonType=Playoffs&ShotClockRange=&StarterBench=&TeamID=0&TwoWay=0&VsConference=&VsDivision='
```

```python
r = requests.get(url=raw_api_url, headers=headers).json()
```

```python
# Check what the response actually contains
print(r.keys())  # This will list the top-level keys of the JSON object
```

```python
result_sets = r['resultSets']
df_rows = result_sets[0]['rowSet']
df_cols_short = result_sets[0]['headers']  # Ensure you're accessing the correct index
df_cols_short
```

```python
df = pd.DataFrame(df_rows, columns=df_cols_short)
df.head(10)
```

#### Looping to get the traditional stats info for some seasons

```python
# Lists of seasons
season_list = [
    '1996-97',
	'1997-98',
	'1998-99',
	'1999-00',
	'2000-01',
	'2001-02',
	'2002-03',
	'2003-04',
	'2004-05',
	'2005-06',
	'2006-07',
	'2007-08',
	'2008-09',
	'2009-10',
	'2010-11',
	'2011-12',
	'2012-13',
	'2013-14',
	'2014-15',
	'2015-16',
	'2016-17',
	'2017-18',
	'2018-19',
	'2019-20',
    '2020-21',
    '2021-22',
]

per_mode = 'Totals'
```

```python
dfs = []
```

```python
for season_id in season_list:
    raw_api_url = 'https://stats.nba.com/stats/leaguedashteamstats?Conference=&DateFrom=&DateTo=&Division=&GameScope=&GameSegment=&Height=&ISTRound=&LastNGames=0&LeagueID=00&Location=&MeasureType=Base&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season=2023-24&SeasonSegment=&SeasonType=Playoffs&ShotClockRange=&StarterBench=&TeamID=0&TwoWay=0&VsConference=&VsDivision='.format(per_mode, season_id)
    r = requests.get(url=raw_api_url, headers=headers).json()
    df = pd.DataFrame(df_rows, columns=df_cols_short)
    df['season_id'] = season_id
    print(season_id)
    dfs.append(df)
```

```python
final_df = pd.concat(dfs, sort=False) 
```

```python
final_df.count()
```

```python
final_df.to_csv('teams_traditional_stats.csv', index=False)
```
